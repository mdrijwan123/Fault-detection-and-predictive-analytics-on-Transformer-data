```python
tsp_unique=df['Sampling_point'].unique()
vol_df=pd.read_excel("C_TRANSFORMER_DETAIL.xlsx")
```

```python
A2_df=pd.DataFrame({
    'Text_id':[],
    'T_site':[],
    'T_plant':[],
    'Sampling_point':[],
    'Rate':[]
})
```

```python
for t_unique in tsp_unique:
    s_date=df[df['Sampling_point']==t_unique]['Sampled_date'].unique()
    samp_date=pd.to_datetime(s_date)
    T=float(abs((samp_date[0]-samp_date[1]).days))
    TDCG1=float(df[(df['Sampled_date']==s_date[0]) & (df['Sampling_point']==t_unique)]['TDCG(ppm)'])
    TDCG2=float(df[(df['Sampled_date']==s_date[1]) & (df['Sampling_point']==t_unique)]['TDCG(ppm)'])
    V=float(vol_df[(vol_df['C_TRANSFORMERS']==t_unique)]['OIL_QUANTITY'])
    Rate=round(pd.to_numeric((abs(TDCG1-TDCG2)*V*(10**-6))/(T)),5)
    Text_id=df[df.Sampling_point==t_unique]['Text_id'].unique()
    T_site=df[df.Sampling_point==t_unique]['T_site'].unique()
    T_plant=df[df.Sampling_point==t_unique]['T_plant'].unique()
    Sampling_point=t_unique 
    
    temp_df=pd.DataFrame({
    'Text_id':[Text_id],
    'T_site':[T_site],
    'T_plant':[T_plant],
    'Sampling_point':[Sampling_point],
    'Rate':[str(Rate*100)+'%']
    })
    A2_df=A2_df.append(temp_df,ignore_index=True)
```

```python
A2_df.to_csv('Annexure2_result.csv')
```
